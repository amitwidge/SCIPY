```python
from scipy import optimize
import matplotlib.pyplot as plt
import numpy as np
from scipy import optimize

x_data = np.linspace(-5, 5, num=50)
y_data = 2.9 * np.sin(1.5 * x_data) + np.random.normal(size=50)

def test_func(x, a, b):
     return a * np.sin(b * x)

# Using optimize to derive the parameters of the model
params, params_covariance = optimize.curve_fit(test_func, x_data, y_data, p0=[2, 2])

print(params)

```

```python
#Curve fitting: temperature as a function of month of the year
months = np.arange(12)

ymax = np.array([39, 41, 43, 47, 49, 51, 45, 38, 37, 29, 27, 25 ])
ymin = np.array([21, 23, 27, 28, 32, 35, 31, 28, 21, 19, 17, 18])


plt.plot(months, ymax, 'ro')
plt.plot(months, ymin, 'bo')
plt.xlabel('Month')
plt.ylabel('Min and max temperature')
```

```python
def yrly_temps(times, avg, ampl, time_offset):
    return(avg + ampl*np.cos((times + time_offset)*2*np.pi/times.max()))

res_max, cov_max = optimize.curve_fit(yrly_temps, months, ymax, [20, 5, 0])
res_min, cov_min = optimize.curve_fit(yrly_temps, months, ymin, [20, 5, 0])


days = np.linspace(0, 12, num = 365)

plt.figure()
plt.plot(months, ymax, 'ro')
plt.plot(days, yrly_temps(days, *res_max), 'r-')
plt.plot(months, ymin, 'bo')
plt.plot(days, yrly_temps(days, *res_min), 'b-')
plt.xlabel('Month')
plt.ylabel('Temperature ($^\circ$C)')


# function to show the plot
# naming the x axis
plt.xlabel('Months')
# naming the y axis
plt.ylabel('Temperature (C)')
# giving a title to my graph
plt.title('Weather')

plt.show()
```
